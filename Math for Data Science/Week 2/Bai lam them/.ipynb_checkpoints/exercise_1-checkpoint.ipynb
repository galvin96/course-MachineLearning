```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('exFuelConsumption.csv')
df.head()
```

```python
df.shape
```

```python
# Lấy 80% dữ liệu để tính
head_df = df[:round(df.shape[0]*0.8)]
head_df
```

```python
# Lấy 20% dữ liệu để test
tail_df = df[-round(df.shape[0]*0.2):]
tail_df
```

```python
X_train = head_df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY']].values
y_train = head_df['CO2EMISSIONS'].values

y_train
```

```python
# Xây dựng hàm tính theo eigen và theo ma trận giả đảo. 
X_test = tail_df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY']].values
y_test = tail_df['CO2EMISSIONS'].values

y_test
# Thêm cột predict vào bộ 20% test là dự đoán dựa trên 3 cột rồi kiểm tra lại độ lệch với cột cuối cùng.
```

```python
def compute_cost(X, y, theta):
    # number of training examples
    m = y.size
    # cost
    j = np.sum((np.dot(X, theta) - y) ** 2)
    j /= (2*m)
    return j
```

```python
def gradient_descent(X, y, theta, alpha, num_inters):
    # number of training examples
    m = y.size
    jHistory = np.empty(num_inters)

    for i in range(num_inters):
        delta = np.dot(np.dot(X, theta) - y, X) / m
        theta -= alpha * delta
        jHistory[i] = compute_cost(X, y, theta)
        if i % 10 == 0:
            print("i: ", i, "theta", theta, "j loss", jHistory[i])
    
    return theta, jHistory
```

```python
# gradient descent max step 
INTERATIONS = 1000
# learning rate
ALPHA = 0.0001
```

```python
theta, jHistory = gradient_descent(X_train, y_train, np.zeros(X_train.shape[1]), ALPHA, INTERATIONS)
print(theta)
# plot J
plt.plot(range(jHistory.size), jHistory, color='g')
plt.xlabel('n'); plt.ylabel('J'); plt.show()
```

```python
y_train_pred = np.dot(X_train, theta)
for a,b in zip(y_train_pred, y_train):
    print(a,b)
```

```python
compute_cost(X_test, y_test, theta)
```

```python
y_test_pred = np.dot(X_test, theta)
for a,b in zip(y_test_pred, y_test):
    print(a,b)
```
