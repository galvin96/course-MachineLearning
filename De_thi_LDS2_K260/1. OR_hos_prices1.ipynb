```python
# Cau 3 Tính tổng số tiền của các bệnh viện vào cột total (0.25 điểm)
df['total'] = df.sum(axis=1)
df.head()
```

```python
# Cau 4 Sắp xếp cột total theo thứ tự giảm dần, chọn ra 30 bệnh viện có mức chi nhiều nhất tạo thành dataframe mới (0.25 điểm)
df.sort_values(by='total', inplace=True, ascending=False)
df_most_total_hospital = df.iloc[:30,]
df_most_total_hospital.head()
```

```python
# Cau 1.5 Tạo bản đồ từ dataframe 30 điểm đã tạo, bản đồ cần tạo chỉ có circlemaker và text là address + amount từ cột total của các điểm (0.5 điểm)
df_most_total_hospital_map = folium.Map(location=[44.0, -120.50], zoom_start=6)
lats = list(df_most_total_hospital.lat)
longs = list(df_most_total_hospital.long)
labels = df_most_total_hospital.address + '</br> Amount: ' + df_most_total_hospital.total.astype(str)

hospitals = folium.map.FeatureGroup()
for lat, long, label in zip(lats, longs, labels):
    hospitals.add_child(
        folium.CircleMarker(
            [lat, long],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='green',
            fill_opacity=0.6,
            popup=label
        )
    ).add_to(df_most_total_hospital_map)
df_most_total_hospital_map
```
