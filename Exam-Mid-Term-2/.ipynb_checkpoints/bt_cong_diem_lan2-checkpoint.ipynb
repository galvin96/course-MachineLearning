```python
import pandas as pd
```

```python
weather = pd.read_csv('weather.csv')
weather.head()
```

```python
# Dùng Pandas vẽ biểu đồ dạng Box của 3 cột 'TMIN', 'TAVG', 'TMAX' và tiêu đề là 'Biểu đồ thời tiết'
weather[['TMIN', 'TAVG', 'TMAX']].plot.box(title='Biểu đồ thời tiết')
```

```python
# Thêm 1 cột dữ liệu 'TDIFF' là sự chênh lệch dữ liệu 'TMAX' - 'TMIN'
weather['TDIFF'] = weather['TMAX'] - weather['TMIN']
weather.head()
```

```python
# Dùng Pandas vẽ biểu đồ histogram của dữ liệu 'TDIFF' với 30 bins và giới hạn trục x là 0 và lưu biểu đồ vào tập tin "tdiff.png"
hist = weather.TDIFF.plot.hist(bins=30, xlim=(0))
fig = hist.get_figure()
fig.savefig('tdiff.png')
```

```python
# Tạo một DataFrame 'bad_weather' mới từ các cột 'WT01' đến 'WT22'
weather_columns = weather.columns[weather.columns.str.startswith('WT')]
bad_weather = weather[weather_columns]
bad_weather
```

```python
# Tạo một cột mới 'bad_weather' trên DataFrame 'weather' bằng cách tính tổng theo cột trên DataFrame 'bad_weather'
weather['bad_weather'] = bad_weather.sum(axis=1)
weather
```

```python
# Điền các giá trị còn thiếu trên cột 'bad_weather' là 0 và chuyển kiểu dữ liệu của cột là int
weather.bad_weather.fillna(0)
weather.astype({'bad_weather': 'int32'}).dtypes
```

```python
# Dùng matplotlib vẽ và lưu (trong tập tin "bad_weather.png") biểu đồ histogram của cột dữ liệu 'bad_weather'. Thiết lập số bins là 10, tiêu đề là 'Thời tiết', màu là 'green'
import matplotlib.pyplot as plt
plt.hist(weather.bad_weather, bins=10)
plt.title('Thời tiết', color='g', fontsize=14)
plt.savefig('bad_weather.png')
```

```python
# Câu 2:
# Cho dữ liệu lưu trong tập tin "movies.csv"
# Thiết lập các thông số cho các biểu đồ: pallete = 'Set2', style = 'whitegrid'
movies = pd.read_csv('movie.csv')
pallete = 'Set2'
style = 'whitegrid'
movies.head()
```

```python
# Vẽ biểu đồ joinplot cho 2 cột dữ liệu x = 'CriticRating', y='AudienceRating'
import seaborn as sb
movies.columns = ['Film','Genre','CriticRating', 'AudienceRating', 'BudgetMillions', 'Year']
sb.jointplot(x='CriticRating', y='AudienceRating', data=movies)
```

```python
# Vẽ biểu đồ displot cho cột dữ liệu 'AudienceRating' với số bins là 15 và tiêu đề là 'Audience Rating'
sb.displot(data=movies, x='AudienceRating', bins=152, kde=True).set(title='Audience Rating')
```

```python
# Vẽ biểu đồ lmplot cho 2 cột dữ liệu x ='CriticRating', y= 'AudienceRating' với hue = 'Genre'
sb.lmplot(data=movies, x ='CriticRating', y= 'AudienceRating', hue='Genre')
```

```python
# Tạo 1 subplot 1 dòng 2 cột, với kích thước (15,6) chia sẽ trục x và y. 
# Vẽ 2 biểu đồ boxplot và violinplot cho 2 cột dữ liệu x ='Genre', y = 'CriticRating' sử dụng pallete màu là 'PRGn'
fig, axs = plt.subplots(1, 2, figsize=(15,6))
d = movies.groupby('Genre').sum('CriticRating')
sb.boxplot(x=movies.Genre, y=movies.CriticRating, palette='PRGn', ax=axs[0])
sb.violinplot(x=movies.Genre, y=movies.CriticRating, palette='PRGn', ax=axs[1])
```
