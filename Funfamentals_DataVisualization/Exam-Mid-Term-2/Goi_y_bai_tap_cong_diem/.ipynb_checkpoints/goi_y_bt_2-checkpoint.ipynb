```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

1. Hãy đọc dữ liệu và lưu vào trong DataFrame 'movies'

```python
movie = pd.read_csv('./movie.csv')
movie.head()
```

2. In một số thông tin tổng quát về dữ liệu. In 4 dòng đầu tiên của DataFrame

```python
movie.info()
```

```python
movie.describe()
```

3. Cho biết số lượng dòng, cột của DataFrame

```python
print('Number of Rows:', movie.shape[0])
print('Number of Columns:', movie.shape[1])
```

4. Hãy cho biết dữ liệu có chứa các giá trị 'NULL' không?

```python
count_false = 0
count_true = 0
count_true = movie.isnull().any().sum()
count_false = movie.isnull().any().shape[0]
print('Number of null value:', count_true)
print('Number of non-null value:', count_false)
```

5. Đổi tên của các cột dữ liệu thành: 'Film','Genre','CriticRating', 'AudienceRating', 'BudgetMillions', 'Year'

```python
movie = movie.rename(columns={'Rotten Tomatoes Ratings %': 'CriticRating', 'Audience Ratings %': 'AudienceRating', 'Budget (million $)': 'BudgetMillions', 'Year of release': 'Year'})
movie.head()
```

6. Thiết lập các thông số cho các biểu đồ: pallete = 'Set2', style = 'whitegrid'

```python
# sns.set()
pallete = 'Set2'
style = 'whitegrid'
```

6. Vẽ biểu đồ joinplot cho 2 cột dữ liệu x = 'CriticRating', y='AudienceRating'

```python
sns.jointplot(data=movie, x='CriticRating', y='AudienceRating')
sns.set_palette(pallete)
sns.set_style(style)
plt.show()
```

7. Vẽ biểu đồ displot cho cột dữ liệu 'AudienceRating' với số bins là 15 và tiêu đề là 'Audience Rating'

```python
sns.distplot(movie['AudienceRating'], bins=15)
plt.title('Audience rating', loc='center')
plt.show()
```

8. Vẽ biểu đồ lmplot cho 2 cột dữ liệu x ='CriticRating', y= 'AudienceRating' với hue = 'Genre'

```python
sns.lmplot(data=movie, x='CriticRating', y='AudienceRating', hue='Genre')
plt.show()
```

9. Tạo 1 subplot 1 dòng 2 cột, với kích thước (15,6) chia sẽ trục x và y.
   Vẽ 2 biểu đồ boxplot và violinplot cho 2 cột dữ liệu x ='Genre', y = 'CriticRating' sử dụng pallete màu là 'PRGn'

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
plt.figure(figsize=(15, 6))
a = sns.boxplot(data=movie, x='Genre', y='CriticRating', palette='PRGn', ax=ax1)
b = sns.violinplot(data=movie, x='Genre', y='CriticRating', palette='PRGn', ax=ax2)
a.set_xticklabels(ax1.get_xticklabels(),rotation=45)
b.set_xticklabels(ax2.get_xticklabels(),rotation=45)
plt.show()
```
