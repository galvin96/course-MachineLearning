```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
weather=pd.read_csv('Tai_nguyen_cung_cap/weather.csv')
weather.head()
```

```python
weather.tail()
```

```python
weather.info()
```

```python
weather.describe()
```

```python
# 2.	Đổi cột timestamp lại thành thời gian đọc được (0.25 điểm)
weather['timestamp']=pd.to_datetime(weather['timestamp'])
weather
```

```python
# 3.	Thực hiện vẽ chart như hình vẽ và nhận xét biểu đồ (0.5 điểm)
fig = plt.figure()
plt.plot('timestamp', 'Hà Nội Temperature [2 m elevation corrected]', data=weather, label='Temperature')
plt.plot('timestamp', 'Hà Nội Wind Speed [10 m]', data=weather, label='Wind Speed')
plt.xticks(rotation=45)
plt.ylabel('Values')
plt.legend()
# weather[['Hà Nội Temperature [2 m elevation corrected]', 'Hà Nội Wind Speed [10 m]']].plot(legend=['Temperature', 'Wind Speed'])
```
