```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
data=pd.read_csv('Tai_nguyen_cung_cap/banh_keo_ratings.csv')
data.head()
```

```python
words = " ".join(data.comment.values)
wordcloud = WordCloud(max_words=500, background_color="black").generate(words)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
mask = np.array(Image.open("Tai_nguyen_cung_cap/shopee.png"))
wordcloud = WordCloud(max_words=500, background_color="black", mask=mask).generate(words)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
