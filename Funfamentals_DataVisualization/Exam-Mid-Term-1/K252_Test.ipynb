## Câu 2:
Cho dữ liệu lưu trong tập tin "movies.csv"
1. Hãy đọc dữ liệu và lưu vào trong DataFrame 'movies'
2. In một số thông tin tổng quát về dữ liệu. In 4 dòng đầu tiên của DataFrame
3. Cho biết số lượng dòng, cột của DataFrame 
4. Hãy cho biết dữ liệu có chứa các giá trị 'NULL' không? 
5. Đổi tên của các cột dữ liệu thành: 'Film','Genre','CriticRating', 'AudienceRating', 'BudgetMillions', 'Year'

```python
import pandas as pd
```

```python
# 1. Hãy đọc dữ liệu và lưu vào trong DataFrame 'weather':
weather = pd.read_csv('weather.csv')
```

```python
# 2. In một số thông tin tổng quát về dữ liệu:
print(weather.info())

# In 4 dòng đầu tiên của DataFrame:
weather.head(4)
```

```python
# 3. Cho biết số lượng dòng, cột của DataFrame:
print("Số dòng:\t", weather.shape[0])
print("Số cột:\t", weather.shape[1])
```

```python
# 4. Hãy cho biết dữ liệu có chứa các giá trị 'NULL' không?

print('Dữ liệu {} chứa các giá trị "NULL".'.format('Có' if weather.isnull().any().any() else 'Không'))
```

```python
# 5. Xóa các cột 'WT10', 'WT11', 'WT15', 'WT19':

weather = weather.drop(columns=['WT10', 'WT11', 'WT15', 'WT19'])
```

```python
# In 3 dòng đầu tiền và 3 dòng cuối cùng của DataFrame

print("3 dòng đầu:\t", weather.head(3))
print("3 dòng cuối:\t", weather.tail(3))
```

```python
# Dùng Pandas vẽ biểu đồ dạng Box của 3 cột 'TMIN', 'TAVG', 'TMAX' và tiêu đề là 'Biểu đồ thời tiết'
import matplotlib.pyplot as plt

axes = weather[['TMIN', 'TAVG', 'TMAX']].boxplot()
fig = axes.get_figure()
fig.suptitle('Biểu đồ thời tiết', fontsize=14, fontweight='bold')
```

```python
# Thêm 1 cột dữ liệu 'TDIFF' là sự chênh lệch dữ liệu 'TMAX' - 'TMIN'
weather['TDIFF'] = weather['TMAX'] - weather['TMIN']
```

```python
# Dùng Pandas vẽ biểu đồ histogram của dữ liệu 'TDIFF'
weather['TDIFF'].plot.hist()
```

```python
# Tạo một DataFrame 'bad_weather' mới từ các cột 'WT01' đến 'WT22'
weather_columns = weather.columns[weather.columns.str.startswith('WT')]
bad_weather = weather[weather_columns]
bad_weather
```

```python
# Tạo một cột mới 'bad_weather' trên DataFrame 'weather' bằng cách tính tổng theo cột trên DataFrame 'bad_weather'
weather['bad_weather'] = bad_weather.sum(axis=1)
weather
```

```python
# Điền các giá trị còn thiếu trên cột 'bad_weather' là 0 và chuyển kiểu dữ liệu của cột là int
weather.bad_weather.fillna(0)
weather.astype({'bad_weather': 'int32'}).dtypes
```

```python
# Lưu dữ liệu trong dataframe 'weather' thành tập tin 'weather.xlsx' với sheetname là 'weather'.
weather.to_excel('weather.xlsx', sheet_name='weather')
```

```python
# Cho dữ liệu lưu trong tập tin "movies.csv"

# 1. Hãy đọc dữ liệu và lưu vào trong DataFrame 'movies'
movies = pd.read_csv('movie.csv')
```

```python
# 2. In một số thông tin tổng quát về dữ liệu. In 4 dòng đầu tiên của DataFrame
print(movies.info())
movies.head(4)
```

```python
# 3. Cho biết số lượng dòng, cột của DataFrame
print("Số dòng:\t", movies.shape[0])
print("Số cột:\t", movies.shape[1])
```

```python
# 4. Hãy cho biết dữ liệu có chứa các giá trị 'NULL' không?
print('Dữ liệu {} chứa các giá trị "NULL".'.format('Có' if movies.isnull().any().any() else 'Không'))
```
