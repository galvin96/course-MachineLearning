```python
import folium
```

```python
world_map = folium.Map(location=[10.762622, 106.660172], zoom_start=10)
world_map
```

```python
import pandas as pd
```

```python
df = pd.read_excel('HCMC_location.xlsx', sheet_name='Location')
df.head()
```

```python
district  = folium.map.FeatureGroup()
```

```python
for lat, long, label in zip(df.Latitude, df.Longitude, df.Name):
    district.add_child(
    folium.CircleMarker(
    [lat, long],
    radius= 5,
    color = 'yellow',
    fill=True,
    fill_color='red',
    fill_opacity=0.6,
    popup = label
    ))
world_map.add_child(district)
```

```python
df = pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=20, skipfooter=2)
df.head()
```

```python
df.rename(columns={'OdName': 'Country'}, inplace=True)
df.head()
```

```python
df['Total'] = df.iloc[:, 9:].sum(axis=1)
df.head()
```

```python
folium.Choropleth(
    geo_data='world-countries.json',
    data=df,
    columns=['Country', 2008],
    key_on='feature.properties.name',
    fill_color='YlOrRd').add_to(world_map2)
world_map2
```
