```python
# 9. Tìm giá trung bình (mean) của trestbps giữa slope và restecg trong tập dataset. Sử dụng group by của slope và restecg để tính (0.5 điểm).
heart[['slope', 'restecg', 'trestbps']].groupby(['slope', 'restecg']).mean()

# heart.groupby(['slope', 'restecg']).mean()
```

```python
figure, axes = plt.subplots(1, 2)
sns.regplot(
    data=heart,
    x='age',
    y='thalach',
    ax=axes[0]
)
sns.regplot(
    data=heart,
    x='age',
    y='trestbps',
    color='y',
    ax=axes[1]
)

# Tuổi càng cao thì tỉ lệ mắc thalach càng giảm,
# Tuổi càng cao thì tỉ lệ mắc trestbps càng tăng
```

```python
d = heart.groupby('age').size().reset_index(name='count').sort_values('count', ascending=False).head()
d.plot.bar(
    'age',
    grid=False,
    legend=False
)
plt.xlabel('Tuổi')
plt.ylabel('Số lượt khảo sát')
plt.title('Số lượng tuổi được khảo sát nhiều nhất ')
```

```python
df_gender = heart.groupby('sex').size().reset_index(name='count')
print(df_gender)
df_gender['count'].plot.pie(autopct='%.2f%%')
plt.legend(['Female', 'Male'])
plt.title('Tỉ lệ Nam/Nữ')
```
