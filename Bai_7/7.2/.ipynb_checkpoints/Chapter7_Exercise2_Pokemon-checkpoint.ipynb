```python
# Câu 1: Vẽ scatterplot: Phân loại với các điểm không chồng chéo, so sánh Attack (trục x) & Defense (trục y)
# sort dữ liệu theo 'Attack', lấy 100 dòng đầu
df = df.sort_values('Attack', ascending=False)
```

```python
plt.figure(figsize=(10,8))
fig = sb.swarmplot(
    x='Attack',
    y='Defense',
    data=df.head(100)
)
plt.xticks(rotation=45 )
```

```python
# Câu 2: Vẽ scatterplot: So sánh Attack (trục x) & Defense (trục y), có sử dụng thêm regression line với lmplot()
sb.lmplot(
    x='Attack',
    y='Defense',
    data=df,
)
```

```python
# Câu 3: Vẽ regression plot: thể hiện mối quan hệ tuyến tính giữa Attack (trục x) & Speed(trục y)
plt.figure(figsize=(8, 6))
sb.regplot(
    x='Attack',
    y='Speed',
    data=df,
    color='r',
    marker='^'
)
```

```python
# Câu 7: Vẽ barplot: Nhóm dữ liệu theo Type 1, cho biết mỗi type có mấy mẫu, sắp tăng dần
grouped = df[['Type 1', 'Name']].groupby('Type 1').count()
sorted = grouped.sort_values('Name')

plt.figure(figsize=(8, 6))
sb.barplot(x=list(grouped.index), y='Name', data=sorted)
plt.xticks(rotation=45)
```

```python
# Câu 8: Vẽ pairplot: thể hiện mối quan hệ giữa "Attack", "Defense" theo từng 'Type 1', sử dụng lại dữ liệu của câu 5
grouped_8 = df[['Type 1', 'Attack', 'Defense']].groupby('Type 1', as_index=False).sum(['Attack', 'Defense'])
grouped_8
sb.pairplot(
    data=df,
    vars=['Attack', 'Defense'],
    palette='bright',
    diag_kind='kde',
    kind='reg',
    hue='Type 1',
    aspect=2,
    height=4
)
```

```python
# Câu 9: Vẽ JointGrid: gồm sb.regplot, sb.distplot để thể hiện mối quan hệ giữa "Attack", "Defense"
sb.jointplot(
    data=df[['Attack', 'Defense']],
    x='Attack',
    y='Defense',
    kind='reg',
    height=8
)
```

```python
# Câu 10: Vẽ JointPlot: thể hiện mối quan hệ giữa 'HP' và 'Speed' dạng hex
sb.jointplot(
    data=df[['HP', 'Speed']],
    x='HP',
    y='Speed',
    kind='hex',
    height=8,
    color='r'
)
```
